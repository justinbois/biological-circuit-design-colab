```python
import numpy as np
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
k = np.logspace(-3, 3, 200)
n = 1
x = 1
y = ((k + x) / (1+x))**n
p = bokeh.plotting.figure(plot_width=400, plot_height=300, x_axis_type='log', y_axis_type='log')
p.line(k, y)
bokeh.io.show(p)
```
